```python
# Step 3: Load and Merge Data

# Replace these filenames with the exact names of your uploaded files
files_list = [
    "_Saturday STEM Academy Registration Fall 2025 (Responses) - Form Responses 1.csv",
    "Saturday STEM Academy Registration Spring 2025 (Responses) - Form Responses 1.csv",
    "Saturday STEM Academy Registration Fall 2024 (Responses) - Form Responses 1.csv",
    "_Saturday STEM Academy Registration Fall 2023 (Responses) - Form Responses 1.csv"
]

# Read each CSV and tag its source
dfs = []
for f in files_list:
    df = pd.read_csv(f)
    df["source_file"] = f  # Add a column to track which file it came from
    dfs.append(df)

# Merge all into one DataFrame
data = pd.concat(dfs, ignore_index=True)

# Preview the merged data
print("Combined shape:", data.shape)
data.head()

```

Grade Level
Purpose: Age/grade distribution of participants.

```python
import numpy as np

def clean_grade_level(grade):
    grade_str = str(grade).strip().lower()
    if grade_str == 'homeschool':
        return grade_str
    try:
        # Check if it's a number
        num_grade = pd.to_numeric(grade_str)
        # Only keep integer grade levels (e.g., 5, not 5.5 or other non-integer floats)
        if num_grade == int(num_grade):
            return int(num_grade)
        else:
            return np.nan # Not an integer grade
    except ValueError:
        return np.nan # Not a number and not 'homeschool'

# Apply the cleaning function to create a new column
data['grade_level_cleaned'] = data['grade_level'].apply(clean_grade_level)

# Get unique valid grades and separate them for sorting
unique_cleaned_grades = data['grade_level_cleaned'].dropna().unique().tolist()

numeric_part = [g for g in unique_cleaned_grades if isinstance(g, int)]
string_part = [g for g in unique_cleaned_grades if isinstance(g, str)]

numeric_part.sort()
plot_order = [str(g) for g in numeric_part] + string_part

# Plot using the cleaned column and the custom order
plt.figure(figsize=(10, 6))
sns.countplot(x='grade_level_cleaned', data=data, order=plot_order, palette='viridis')
plt.title("Grade Level Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Grade Level", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import re
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid", context="talk")

# Helper: detect class-choice columns per file (handles varied naming)
def get_class_columns(df):
    cols = []
    for c in df.columns:
        lc = c.lower()
        if ("choose" in lc and "class" in lc) or re.fullmatch(r"column[_\s]*\d+", lc):
            cols.append(c)
    return cols

# Helper: clean and normalize class labels
def clean_choice(x):
    if pd.isna(x):
        return None
    s = str(x).strip().rstrip(",")  # trim spaces and trailing commas
    if s == "" or s.lower() == "nan":
        return None
    # Minimal normalization map for known duplicates
    normalized = {
        "storytime with robots": "Storytime with Robots",
        "storytime with robots,": "Storytime with Robots",
        "animal engineers": "Animal Engineers",
        "animal engineers,": "Animal Engineers",
        "stemtastic fables and fairytales": "STEMtastic Fables and Fairytales",
        "chemistreat": "Chemistreat",
    }
    key = s.lower()
    return normalized.get(key, s)
```

```python
# Ensure source_file exists and columns are consistent
assert "source_file" in data.columns, "source_file column missing — please add it when loading each CSV."

for file_name in sorted(data["source_file"].unique()):
    df_f = data[data["source_file"] == file_name].copy()
    class_cols = get_class_columns(df_f)

    print(f"\n=== {file_name} ===")
    print(f"Detected class-choice columns: {class_cols if class_cols else 'None'}")

    if not class_cols:
        print("No class-choice columns found. Skipping plot.")
        continue

    # Clean class columns
    choices = df_f[class_cols].applymap(clean_choice)

    # Count students who did NOT choose any class (all class columns empty/None)
    no_choice_mask = choices.apply(
        lambda row: all(x is None for x in row), axis=1
    )
    did_not_choose = int(no_choice_mask.sum())
    chose_any = int(len(df_f) - did_not_choose)

    # Long format of all chosen classes, drop None
    long_choices = (
        choices.melt(value_name="class_choice")["class_choice"]
        .dropna()
    )

    # Plot class popularity per file
    counts = long_choices.value_counts()

    plt.figure(figsize=(10, 6))
    sns.barplot(y=counts.index, x=counts.values, palette="Blues_r")
    plt.title(f"Class Choices Distribution — {file_name}", fontsize=16, fontweight="bold")
    plt.xlabel("Count", fontsize=14)
    plt.ylabel("Class", fontsize=14)
    plt.tight_layout()
    plt.show()

    # Plot chose vs. did not choose for this file
    status = pd.Series({"Chose a class": chose_any, "Did not choose": did_not_choose})

    plt.figure(figsize=(6, 3.5))
    sns.barplot(y=status.index, x=status.values, palette=["green", "red"])
    plt.title(f"Class Selection Status — {file_name}", fontsize=14, fontweight="bold")
    plt.xlabel("Number of Students", fontsize=12)
    plt.ylabel("")
    plt.tight_layout()
    plt.show()

    # Quick text summary
    print(f"Total rows: {len(df_f)} | Chose a class: {chose_any} | Did not choose: {did_not_choose}")
    if len(counts) > 0:
        top = counts.head(5)
        print("Top classes:")
        for cls, cnt in top.items():
            print(f" - {cls}: {cnt}")
```
