A company started to invest in digital marketing as a new way of their product
promotions. For that they collected data and decided to carry out a study on it.
‚óè The company wishes to clarify whether there is any increase in sales after
stepping into digital marketing.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
data = pd.read_csv('/content/Sales_add.csv')
# print(data.head())
print(data.columns)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Month'], data['Sales_After_digital_add(in $)'], marker='o', color='b')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sales_before = data['Sales_before_digital_add(in $)']
sales_after = data['Sales_After_digital_add(in $)']
t_statistic, p_value = ttest_ind(sales_before, sales_after)
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference in sales before and after digital marketing.")
else:
    print("There is no statistically significant difference in sales before and after digital marketing.")
```

The company needs to check whether there is any dependency between
thefeatures ‚ÄúRegion‚Äù and ‚ÄúManager‚Äù.


```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Region'], data['Manager'])
chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
alpha = 0.05
if p_val < alpha:
    print("There is a significant dependency between Region and Manager.")
else:
    print("There is no significant dependency between Region and Manager.")
```
