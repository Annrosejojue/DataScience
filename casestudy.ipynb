```python
import numpy as np
original_array = np.array([[8, 9, 10], [11, 12, 13]])
reshaped_array = original_array.reshape(3, 2)
print("Original Array:")
print(original_array)
print("\nReshaped Array:")
print(reshaped_array)
```

```python
original_array = np.array([[8, 9], [10, 11], [12, 13]])
sliced_array = original_array[:, [1]].flatten()
print("Original Array:")
print(original_array)
print("\nSliced Array:")
print(sliced_array)
```

```python
file_path = "/diabetes_model.csv"
df = pd.read_csv(file_path)
subset = df.loc[45:60, ['Glucose', 'Age']]
print(subset)
```

```python
above_120_count = df[df['BloodPressure'] > 120].shape[0]
print(f"Number of patients with blood pressure above 120: {above_120_count}")
```

```python
min_insulin = df['Insulin'].min()
max_insulin = df['Insulin'].max()
print(f"Minimum insulin level: {min_insulin}")
print(f"Maximum insulin level: {max_insulin}")
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], color='skyblue', edgecolor='black')
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['BMI'], df['Glucose'], color='blue')
plt.title('Scatter Plot of BMI vs Glucose')
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```
