```python
df['Work-Life Balance'].fillna(df['Work-Life Balance'].mode()[0], inplace=True)
df['Performance Rating'].fillna(df['Performance Rating'].mode()[0], inplace=True)
df['Number of Promotions'].fillna(df['Number of Promotions'].mean(), inplace=True)
df['Education Level'].fillna(df['Education Level'].mode()[0], inplace=True)
df['Marital Status'].fillna(df['Marital Status'].mode()[0],inplace=True)
df['Number of Dependents'].fillna(df['Number of Dependents'].mean(), inplace=True)
df['Job Level'].fillna(df['Job Level'].mode()[0], inplace=True)
df['Company Size'].fillna(df['Company Size'].mode()[0], inplace=True)
df['Remote Work'].fillna(df['Remote Work'].mode()[0],inplace=True)
df['Employee Recognition'].fillna(df['Employee Recognition'].mode()[0], inplace=True)
```

```python
#Print all the columns and their Unique Values
for column in df.columns:
  if df[columns].dtype == 'object':
    print(str(column)+ '  :'+ str(df[column].unique()))
    print(df[column].value_counts())
    print('------------------------------------------------------------')
```

```python
#Remove Unnecessary columns
df.drop('Employee ID',axis=1,inplace=True)
```

```python
df=pd.get_dummies(df)
```

```python
correlation_matrix = df.corr()

# Display the correlation matrix
correlation_matrix
```

```python
#Transform the data
#Transform Numerical col into Non Numerical Column
from sklearn.preprocessing import LabelEncoder

for column in df.columns:
  if df[column].dtype==np.number:
    continue
    df[column]=LabelEncoder().fit_transform(df[column])
```

```python
#Create a New Column
df['Age_year']=df['Age']
```

```python
df.drop('Age',axis=1,inplace=True)
```

```python
df
```

```python
#Split the data
X=df.iloc[:,1:df.shape[1]].values
Y=df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#Use random forest Classifier
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(n_estimators=100,random_state=42)
forest.fit(X_train,Y_train)
```

```python
#get the Accuracy on the Training dataset
forest.score(X_train,Y_train)
```

```python
#Confusion Matrix and and accuracy for the model on the test data
from sklearn.metrics import confusion_matrix,classification_report
Y_pred=forest.predict(X_test)
cm=confusion_matrix(Y_test,Y_pred)
```

```python
TN=cm[0][0]
TP=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print('Testing Accuracy=',(TP+TN)/(TP+TN+FN+FP))
```
